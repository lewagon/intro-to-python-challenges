```python
def update_grades(file):
    print(f"Starting with {file}")
    
    # 1. importing pandas
    import pandas as pd
    
    # 2. reading the Excel file
    df = pd.read_excel('grades.xlsx')
    
    # 3. Adding an average grade per subject
    df = df.append(df.mean(axis='rows'), ignore_index=True)
    
    # 4. Adding an average grade per student
    df['Average Grade'] = df.mean(axis='columns')
    
    # 5. Marking failing grades
    def red_grade_if_failing(grade):
        if grade < 7:
            return ("color: red;")
        
    sdf = df.style.applymap(red_grade_if_failing, subset='Average Grade')
    
    # 6. Marking top performers in each subject
    def green_grade_if_max(subject):
        return ['background-color: green; color: white;' if grade == max(subject) else None for grade in subject]
    
    sdf.apply(green_grade_if_max, axis='rows', subset=['Python Fundamentals', 'Statistics', 'Data Visualization', 'Presentation'])
    
    # 7. Saving back to the file
    sdf.to_excel(file, index=False)
    print(f"{file} saved!")
```

```python
update_grades('styled_grades.xlsx')
```
